```python
#imports
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from mpl_toolkits import mplot3d

#https://en.wikipedia.org/wiki/Characteristic_function_(probability_theory)
#Variables


i = np.complex(0,1);
```

```python
#Plot functions

def plot_2d(t,fn, fig_index, y_label):
  plt.figure(fig_index);
  plt.plot(t,fn(t))
  plt.grid()
  plt.xlabel("t")
  plt.ylabel(y_label)

def plot_3d(t,fn, fig_index):
  plt.figure(fig_index)
  ax = plt.axes(projection="3d")
  fnv = fn(t)
  x = np.real(fnv)
  y = np.imag(fnv)
  ax.plot3D(t,x,y,"blue")
  ax.view_init(30,-120)
  plt.title("Chf")


def plot_chf_2d(t,fn, fig_index):
  plt.figure(fig_index)  
  fnv = fn(t)
  x = np.real(fnv)
  y = np.imag(fnv)
  plt.plot(t,x)
  plt.grid()
  plt.xlabel("x")
  plt.ylabel("y")
```

```python
#Normal Distribution

def normal_dist(mu, sigma_sq):  
  t = np.linspace(5, 15, 100)
  u = np.linspace(0,5,250);
  plot_2d(t,lambda t: st.norm.pdf(t,mu,sigma_sq),1,"pdf")
  plot_2d(t,lambda t: st.norm.cdf(t, mu, sigma_sq),2,"cdf") 
  plot_3d(u,lambda u: np.exp(i*mu*u - ((sigma_sq * sigma_sq * u * u)/2.0)),3)

normal_dist(10,1)
normal_dist(9,1)
```

```python
def uniform_dist(a,b):
  t1 = np.linspace(a, b,40)
  n = b-a+1
  plot_2d(t1,lambda t1: st.uniform.pdf(t1),1,"pdf")
  plot_2d(t1,lambda t1: st.uniform.cdf(t1),2,"cdf") 
  plot_chf_2d(t1,lambda t1: (np.exp(i*t1*a) - np.exp(i*t1*(b+1)))/(n*(1-np.exp(i*t1))),3)
#uniform_dist(-1,1)
uniform_dist(-5,5)
```
