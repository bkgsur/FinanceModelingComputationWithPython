```python
import numpy as np
import pandas as pd
```

```python
f = 5
n = 5
np.random.seed(100)

X = np.random.randint(0,2,(n,f))
X[:4,:3]

Y = np.random.randint(0,2,n)
Y[:4];

print(f"Patterns count {2**f}")

fcols = [ f'f{i}' for i in range(0,f)]
fcols

data = pd.DataFrame(X, columns=fcols) # features
data['l'] = Y # labels

data.info()
```

```python
print(data)

grouped = data.groupby(list(data.columns))

print(grouped['l'].size())

freq = grouped['l'].size().unstack(fill_value=0)
freq['sum'] = freq[0] + freq[1]
freq
```
