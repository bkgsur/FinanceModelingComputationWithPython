```python
import numpy as np
import matplotlib.pyplot as plt
import math
import itertools
import scipy.stats as st
```

```python
# Standard Brownian Motion
N = 10 # number of simul
T= 20 #time steps

random_walks = [1,-1] # heads/tails

plt.figure(1)
for i in range(N):
  steps = np.random.choice(random_walks,size=T).T  
  origin = np.zeros(1)
  path = np.concatenate([origin,steps]).cumsum(axis=0) 
  plt.plot(path)
plt.grid()
plt.title("Standard Brownian Motion")
plt.xlabel("t")
plt.ylabel("Move(s)")
plt.show()


```
