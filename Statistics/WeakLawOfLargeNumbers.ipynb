**Weak law of large numbers**

Distribution of sample becomes concentrated around population mean μ as n→∞.



```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
def weak_law(p_dist):
  N=1000
  #Population
  p = p_dist.rvs(N)  
  s_prop = lambda n: np.mean(p[:n+1])
  p_prop = p_dist.mean()
  sample_prop_list = [s_prop(n) for n in range(N)]
  plt.plot(p)
  plt.plot(sample_prop_list, color='r')
  plt.axhline(p_prop, color='g', linestyle='-')
  plt.xlabel('N')
  plt.ylabel("Mean")

  print('Difference between population and sample mean:', sample_prop_list[-1]-p_prop)
```

```python
print("Bernoulli Distribution")
p = 0.7 
p_dist = scipy.stats.bernoulli(p) 
weak_law(p_dist)
```

```python
print("Normal Distribution")
mu = 175 # population mean
sigma_sq = 10  # population variance
p_dist = scipy.stats.norm(175, np.sqrt(10)) 
weak_law(p_dist)
```

```python
print("Uniform Distribution")
p_dist = scipy.stats.uniform()
weak_law(p_dist)
```
