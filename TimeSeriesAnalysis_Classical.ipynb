### Time Series Oil Futures (Brent)

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn')
```

Read Data

```python
brent_futures =  pd.read_csv('/content/data/Brent Oil Futures Historical Data_Monthly.csv', parse_dates=True, index_col= 0)
print("Head:\n",brent_futures.head())
print("Tail:\n",brent_futures.tail()) 
print("Summary Statistics:\n",brent_futures.describe())
```

```python
plt.figure(figsize=(24, 20))
plt.plot(brent_futures.Change_Percent)
plt.ylabel('% Change')
plt.xlabel('Month')
plt.show()
```

```python
seasonal_decompose(brent_futures.Change_Percent, freq=12).plot()
plt.show()
```

## Visualize Montly oil prices

```python
plt.figure(figsize=(24, 20))
plt.plot(brent_futures.Price)
plt.ylabel('% Change')
plt.xlabel('Month')
plt.show()
```
