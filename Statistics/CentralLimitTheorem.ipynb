**Central Limit Theorem: 
Even when a sample is not normally distributed, if you draw multiple samples and take each of their averages, these averages will represent a normal distribution.**

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
def clt(p_dist):
  N= 1000
  p_mean_list = []
  for _ in range(1,5000):
    p_dist = scipy.stats.uniform()
    p = p_dist.rvs(N)  
    p_mean_list.append(np.mean(p))

  p_mean_list.sort()
  mu = np.mean(p_mean_list)
  sigma = np.std(p_mean_list)
  pdf = scipy.stats.norm.pdf(p_mean_list, mu, sigma)
  plt.plot(p_mean_list, pdf) 
  plt.show()
  # normality test
  stat, p = scipy.stats.shapiro(p_mean_list)
  print('Statistics=%.3f, p=%.3f' % (stat, p))
  
  alpha = 0.05
  if p > alpha:
    print('Gaussian (fail to reject H0)')
  else:
    print('Not look Gaussian (reject H0)')
```

```python
print("Bernoulli distribution")
p_dist = scipy.stats.bernoulli(0.7)
clt(p_dist)
```

```python
print("Uniform distribution")
p_dist = scipy.stats.uniform()
clt(p_dist)
```

```python
print("Exponential distribution")
p_dist = scipy.stats.expon()
clt(p_dist)
```
