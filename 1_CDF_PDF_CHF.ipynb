```python
#imports
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from mpl_toolkits import mplot3d

#https://en.wikipedia.org/wiki/Characteristic_function_(probability_theory)
#Variables

t = np.linspace(5, 15, 100)
u = np.linspace(0,5,250);
```

```python
#Plot functions

def plot_2d(fn, fig_index, y_label):
  plt.figure(fig_index);
  plt.plot(t,fn(t))
  plt.grid()
  plt.xlabel("t")
  plt.ylabel(y_label)

def plot_3d(fn, fig_index):
  plt.figure(fig_index)
  ax = plt.axes(projection="3d")
  fnv = fn(u)
  x = np.real(fnv)
  y = np.imag(fnv)
  ax.plot3D(u,x,y,"blue")
  ax.view_init(30,-120)
```

```python
#Normal Distribution

def normal_dist(mu, sigma_sq):  
  plot_2d(lambda t: st.norm.pdf(t,mu,sigma_sq),1,"pdf")
  plot_2d(lambda t: st.norm.cdf(t, mu, sigma_sq),2,"cdf")
  i = np.complex(0,1);
  plot_3d(lambda u: np.exp(i*mu*u - ((sigma_sq * sigma_sq * u * u)/2.0)),3)

normal_dist(10,1)
```
